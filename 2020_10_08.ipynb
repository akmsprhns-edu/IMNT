```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(-10,10,0.1)
```

```python
from numpy import exp
def sigmoid(val):
    return 1.0/(1.0+exp(-val))
def tanh(val):
    return (exp(val)-exp(-val))/(exp(val)+exp(-val))
def step(val):
    return np.where(val < 0, 0, 1)
def lin(val):
    return np.where(val < 0, 0,
                   np.where(val < 1, val, 1))
```

```python
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_step = step(x)
y_lin = lin(x)
plt.figure(dpi=200)
plt.plot(x,y_sigmoid)
plt.plot(x,y_tanh)
plt.plot(x, y_step)
plt.plot(x, y_lin)
plt.show()
```

```python
def percept_fun(x, w):
    x = np.append(x, np.ones((len(x),1)), axis=1)
    net = np.dot(x,w)
    return step(net)
```

```python
import pandas as pd
csvdata = pd.read_csv("dati.csv")
data = csvdata.to_numpy()
w = [1,0.5,-6]
pc = percept_fun(data[:,:2],w)
img = np.zeros(60,70);
plt.scatter(data[:,0], data[:,1], c=data[:, 2], marker='s', s = 300)
plt.scatter(data[:, 0], data[:, 1], c=pc, linewidths= 1, edgecolors='black')
plt.show()
```
