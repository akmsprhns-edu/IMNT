```python
import numpy as np

lc = 0.1  #Apmācības koeficients

class Neirons:
    
    def __init__(self, inSize):
        self.w = np.zeros(inSize)
        self.w0 = 0
    
    def classify(self, x):
        net = np.dot(x, self.w)
        net += self.w0
        y = 0 if net < 0 else 1 
        return y
    
    def learn(self, x, z):
        y = self.classify(x)
        delta = z-y
        if delta != 0:
            for i in range(0, len(self.w)):
                self.w[i]+= lc*delta*x[i]
            self.w0+= lc*delta
```

```python
n1 = Neirons(15);
n2 = Neirons(15);
n3 = Neirons(15);

for eph in range(1000):
    for bi in range(0, len(burti)):
        n1.learn(burti[bi][1],burti[bi][2][0])
        n2.learn(burti[bi][1],burti[bi][2][1])
        n3.learn(burti[bi][1],burti[bi][2][2])

for bi in range(0, len(burti)):
    print(burti[bi][0]+" "+str(n1.classify(burti[bi][1]))+" "+str(n2.classify(burti[bi][1]))+" "+str(n3.classify(burti[bi][1])))
    
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

print(iris)
```

```python
import tensorflow as tf
#  from keras.layers import Dense
#  from keras.models import Sequential

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(3, activation='softmax', input_shape = (len(iris.data[0]), )))

model.compile(tf.optimizers.RMSprop(0.001), loss='mse')

targets = np.zeros((len(iris.target),3))
for ti in range(0, len(iris.target)):
    if iris.target[ti] == 0:
        targets[ti][0] = 1
    elif iris.target[ti] == 1:
        targets[ti][1] = 1
    else:
        targets[ti][2] = 1

        
model.fit(iris.data, targets, epochs=300)

predictions = model.predict(iris.data)

```
